## Task1:

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
df=pd.read_csv('DS1_C4_S7_Car_Data_Challenge.csv')
```

```python
T1=df[['Cylinders','Kerb_Weight','Make','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque']]
```

```python
T1.dropna(inplace=True)
T1
```

## Task2:

```python
df['Kerb_Weight']=df['Kerb_Weight'].str.replace('kg','').str.replace('-','')
df['Kerb_Weight']=pd.to_numeric(df['Kerb_Weight'])
df['Kerb_Weight']=df['Kerb_Weight']*2.2
df[['Kerb_Weight']]
```

```python
df['Power']=df['Power'].str.extract('(\d+)')
df['Power']=pd.to_numeric(df['Power'])
df[['Power']]
```

```python
df['Torque']=df['Torque'].str.extract('(\d+)')
df['Torque']=pd.to_numeric(df['Torque'])
df[['Torque']]
```

## Task3:

```python
T1['Power']=df[['Power']]
T1['Torque']=df[['Torque']]
```

```python
T3_mean=T1.groupby(by='Make')['Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque'].mean()
T3_mean
```

```python
T1_median=T1.groupby(by='Make')['Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque'].median()
T1_median
```

```python
T1_std=T1.groupby(by='Make')['Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque'].std()
T1_std
```

```python
T1_cov=(T1.groupby(by='Make')['Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque'].std()/T1.groupby(by='Make')['Cylinders','Kerb_Weight','Displacement','Fuel_Tank_Capacity_litre','City_Mileage_km_litre','Highway_Mileage_km_litre','Power','Torque'].mean())*100
T1_cov
```

```python
import matplotlib.pyplot as plt
T1.boxplot(by='Make',column='Displacement',figsize=(15,9))
plt.xticks(rotation=90)
plt.show()
```

## Task4:

```python
T4=df[(df['Make']=='Renault')|(df['Make']=='Toyota') |(df['Make']=='Honda')]
```

```python
import seaborn as sns
sns.regplot(x=T4['Displacement'],y=T4['City_Mileage_km_litre'])
plt.title('Relationship between Displacement and City Mileage')
```

###### In the above case more displacement will effect car mileage negatively and less displacement will have positive effect on mileage

```python
sns.regplot(x=T4['Torque'],y=T4['City_Mileage_km_litre'])
plt.title('Relationship between Torque and City Mileage')
```

###### In case of Torque also - More Torque means less mileage and less Torque means more mileage

```python
sns.regplot(x=T4['Power'],y=T4['City_Mileage_km_litre'])
plt.title('Relationship between Power and City Mileage')
```

###### Power also plays the same role in cars mileage- less power means efficient mileage and more power means reduction in mileage.

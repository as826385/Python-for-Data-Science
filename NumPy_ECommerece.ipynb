```python
import numpy as np
```

```python
data=np.genfromtxt('DS1_C4_S1_Shopping_Data_Challenge.csv',delimiter=',',skip_header=1)
```

### Task1:

```python
index=np.where(data[:,3]>80)
details=data[index]
count=np.array(index)
print("The count of xustomer whose spending score is above 80 is: ",count.size)
print()
print(details)
```

### Task2:

```python
index=np.where((data[:,1]>=20)&(data[:,1]<=24))
a=data[index]
print(a)
```

### Task3:

```python
max_sal=np.max(data[:,2])
min_sal=np.min(data[:,2])
index=np.where((data[:,2]==max_sal) | (data[:,2]==min_sal))
ID=data[:,0][index]
print(ID)
```

### Task4:

```python
index=np.where((data[:,1]<45)&(data[:,2]>35)&(data[:,3]>75))
details=data[index]
count=np.array(index)
print("Number of customers that satisfy above conditions are: ",count.size)
print()
print(details)
```
